```python
import pandas as pd
```

## 1. Importación de Pandas

En este paso se importa la librería **Pandas**, la cual es ampliamente utilizada en Python para el análisis y manipulación de datos.

Se utiliza el alias `pd` para simplificar su escritura en el código.

```python
datos = {
    "Nombre": ["Ana", "Luis", "Carlos", "María", "Sofía"],
    "Edad": [22, 25, 30, 28, 21],
    "Ciudad": ["Bogotá", "Medellín", "Cali", "Barranquilla", "Cartagena"],
    "Carrera": ["Ingeniería", "Diseño", "Administración", "Contabilidad", "Psicología"]
}

df = pd.DataFrame(datos)
df
```

## 2. Creación de un DataFrame

En este ejercicio se crea un **DataFrame** a partir de un diccionario de Python.

Cada clave representa una **columna**, mientras que cada lista contiene los valores correspondientes a esa columna.

El DataFrame organiza la información en forma de tabla, lo cual facilita su exploración y análisis.

```python
df.head()
```

## 3. Visualización inicial de datos

Con el método `head()` se visualizan las primeras filas del DataFrame.

Esto es útil para revisar rápidamente si la estructura y el contenido de los datos son correctos.

```python
df.info()
```

## 4. Información general del DataFrame

El método `info()` permite observar información general sobre el DataFrame, como:

- número de filas
- número de columnas
- nombres de las columnas
- tipos de datos
- cantidad de valores no nulos

Este análisis ayuda a comprender mejor la estructura interna de los datos.

```python
df.describe()
```

## 5. Estadísticas descriptivas

El método `describe()` genera un resumen estadístico de las columnas numéricas del DataFrame.

Entre los valores mostrados se encuentran:

- cantidad de datos
- promedio
- desviación estándar
- valor mínimo
- valor máximo

Esto facilita una exploración inicial de la información numérica.

```python
df["Nombre"]
```

## 6. Selección de una columna

Aquí se selecciona únicamente la columna **Nombre** del DataFrame.

Esto permite trabajar específicamente con una sola variable de la tabla.

```python
df[["Nombre", "Edad"]]
```

## 7. Selección de varias columnas

En este caso se extraen las columnas **Nombre** y **Edad**.

Esto resulta útil cuando se desea analizar solamente ciertos datos sin mostrar toda la tabla.

```python
df.loc[0]
```

## 8. Selección de una fila

El método `loc[]` permite seleccionar filas específicas utilizando su índice.

En este ejemplo se accede al primer registro del DataFrame.

```python
df["Salario"] = [1500000, 1800000, 2500000, 2200000, 1600000]
df
```

## 9. Agregar una nueva columna

En este paso se agrega una nueva columna llamada **Salario**.

Esto demuestra que los DataFrames permiten incorporar nueva información de forma sencilla y flexible.

```python
df[df["Edad"] > 24]
```

## 10. Filtrado por condición numérica

Aquí se filtran únicamente las filas donde la edad es mayor a 24 años.

Los filtros son muy útiles para extraer subconjuntos de información dentro de un conjunto de datos más grande.

```python
df[df["Ciudad"] == "Bogotá"]
```

## 11. Filtrado por condición de texto

En este ejemplo se filtran los registros cuya ciudad sea **Bogotá**.

Esto permite localizar información puntual dentro del DataFrame.

```python
df.sort_values("Edad")
```

## 12. Ordenamiento de datos

Con `sort_values()` se pueden ordenar los datos según una columna específica.

En este caso, las personas se ordenan de menor a mayor según la edad.

```python
df["Salario"].mean()
```

## 13. Cálculo del promedio

Con el método `mean()` se calcula el promedio de la columna **Salario**.

Este tipo de operación es útil para realizar análisis estadísticos básicos.

```python
df["Salario"].max()
```

## 14. Valor máximo

Con el método `max()` se obtiene el valor más alto de la columna **Salario**.

Esto permite identificar el valor máximo dentro de un conjunto de datos numéricos.

```python
df.to_csv("personas.csv", index=False)
```

## 15. Exportación del DataFrame

En este paso se exporta el DataFrame a un archivo CSV llamado `personas.csv`.

Esto permite guardar la información para usarla posteriormente en otros análisis o herramientas.

```python
df
```

```python
# ejersicio adicional Mostrar solo las personas con salario mayor a 1.700.000
df[df["Salario"] > 1700000]
```

## 16. Ejercicio adicional

En este ejercicio adicional se filtran las personas cuyo salario es mayor a **1.700.000**.

Esto refuerza el uso de condiciones para seleccionar información específica dentro del DataFrame.
