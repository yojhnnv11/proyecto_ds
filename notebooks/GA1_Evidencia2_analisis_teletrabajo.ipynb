```python
# Importación de librerías necesarias
import pandas as pd
import numpy as np
```

```python
# ============================================================
# CARGA DEL DATASET
# ============================================================
# El archivo se encuentra fuera de la carpeta notebooks,
# por lo tanto se utiliza ../ para subir un nivel en la ruta.
# ============================================================

df = pd.read_csv('../remote_work_burnout_dataset.csv')
```

```python
# Mostrar las primeras filas para confirmar que el archivo cargó correctamente
df.head()
```

```python
# ============================================================
# INSPECCIÓN RÁPIDA DEL DATASET
# ============================================================
# Se revisan las primeras filas para conocer:
# - nombres de columnas
# - tipo de información contenida
# - forma general de los datos
# ============================================================

df.head()
```

```python
# Mostrar las últimas filas del dataset
# Esto ayuda a revisar cómo termina el conjunto de datos
df.tail()
```

```python
# ============================================================
# TAMAÑO DEL DATASET
# ============================================================
# .shape devuelve:
# (cantidad de filas, cantidad de columnas)
# Esto permite conocer el tamaño del conjunto de datos
# ============================================================

df.shape
```

```python
# Mostrar los nombres de las columnas
# Esto ayuda a comprender la estructura del dataset
df.columns
```

```python
# Mostrar los tipos de datos de cada columna
# Esto permite saber qué variables son numéricas y cuáles categóricas
df.dtypes
```

```python
# ============================================================
# DIAGNÓSTICO PRINCIPAL DEL DATASET
# ============================================================
# .info() permite observar:
# - número total de registros
# - número de columnas
# - cantidad de valores no nulos
# - tipo de dato de cada variable
# ============================================================

df.info()
```

```python
# ============================================================
# REVISIÓN DE VALORES NULOS
# ============================================================
# Esto permite detectar si hay datos faltantes por columna
# y evaluar posibles problemas de calidad
# ============================================================

df.isnull().sum()
```

```python
# Ordenar los valores nulos de mayor a menor
# Esto ayuda a identificar rápidamente qué columnas requieren más atención
df.isnull().sum().sort_values(ascending=False)
```

```python
# ============================================================
# REVISIÓN DE DUPLICADOS
# ============================================================
# Esto permite saber si existen filas repetidas dentro del dataset
# Los duplicados pueden afectar el análisis si no se controlan
# ============================================================

df.duplicated().sum()
```

```python
# ============================================================
# RESUMEN ESTADÍSTICO DE VARIABLES NUMÉRICAS
# ============================================================
# .describe() permite obtener:
# - conteo
# - promedio
# - desviación estándar
# - valor mínimo
# - percentiles
# - valor máximo
# ============================================================

df.describe()
```

```python
# ============================================================
# RESUMEN ESTADÍSTICO DE VARIABLES CATEGÓRICAS
# ============================================================
# .describe(include='object') permite revisar:
# - cantidad de valores únicos
# - valor más frecuente
# - frecuencia del valor más repetido
# ============================================================

df.describe(include='object')
```

```python
# ============================================================
# REVISIÓN DE VALORES ÚNICOS EN COLUMNAS CATEGÓRICAS
# ============================================================
# Esto ayuda a entender mejor las categorías presentes en el dataset
# y detectar posibles inconsistencias en nombres o formatos
# ============================================================

for col in df.select_dtypes(include='object').columns:
    print(f"\nColumna: {col}")
    print(df[col].unique()[:10])  # muestra hasta 10 valores únicos
```

```python
# Cantidad de valores únicos por columna
# Esto ayuda a identificar variables con alta o baja diversidad de datos
df.nunique()
```

## 6. Tipos de datos identificados

Durante el diagnóstico se observó la presencia de dos tipos principales de variables:

### Variables numéricas
Corresponden a columnas relacionadas con mediciones cuantitativas, como por ejemplo:

- Horas de trabajo
- Indicadores de productividad

### Variables categóricas
Corresponden a variables descriptivas, como por ejemplo:

- Modalidad de trabajo

## 7. Identificación de valores nulos

Uno de los aspectos clave dentro del diagnóstico de calidad de datos fue la detección de valores faltantes.

Para ello, se utilizó el método `.isnull().sum()`, el cual permitió conocer la cantidad de datos nulos presentes en cada columna.

La presencia de valores faltantes puede representar un problema importante en el análisis, ya que puede afectar:

- cálculos estadísticos,
- visualizaciones,
- correlaciones,
- e interpretación de resultados.

Por esta razón, identificar los valores nulos desde el inicio permite tomar decisiones posteriores sobre imputación, eliminación o tratamiento de dichos registros.

## 8. Revisión de duplicados

También se realizó una validación de registros duplicados mediante el uso de `.duplicated().sum()`.

Este paso es importante porque los datos duplicados pueden distorsionar el análisis y generar conclusiones erróneas, especialmente cuando se trabaja con encuestas o respuestas repetidas.

La revisión de duplicados permite asegurar una mayor confiabilidad del conjunto de datos antes de aplicar transformaciones o análisis descriptivos.

- desviación estándar,

## 10. Resumen estadístico de variables categóricas
