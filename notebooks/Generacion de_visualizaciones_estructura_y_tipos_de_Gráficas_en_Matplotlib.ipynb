````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Configuración opcional para notebook
%matplotlib inline
````

````python
# Datos sintéticos para replicar ejercicios de Matplotlib
np.random.seed(42)

x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)

data = np.random.normal(50, 10, 200)
data2 = np.random.normal(60, 15, 200)

x_scatter = np.random.rand(100) * 10
y_scatter = 2 * x_scatter + np.random.randn(100) * 2

categorias = ['Remoto', 'Híbrido', 'Presencial']
satisfaccion = [78, 65, 58]
productividad = [82, 74, 69]
burnout = [45, 52, 61]
````

## A. Estructura de una Gráfica

En esta sección se replican ejercicios relacionados con:
- creación de figuras,
- ejes,
- títulos,
- etiquetas,
- leyendas,
- anotaciones.

````python
plt.figure(figsize=(8,5))
plt.plot(x, y, label='Seno')
plt.title('Ejemplo con plt.figure()')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.grid(True)
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(x, y, label='Seno')
ax.set_title('Ejemplo con plt.subplots()')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.legend()
ax.grid(True)
plt.show()
````

````python
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(1,1,1)
ax.plot(x, y, label='Seno')
ax.set_title('Ejemplo con add_subplot()')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.legend()
ax.grid(True)
plt.show()
````

````python
plt.figure(figsize=(8,5))
plt.plot(x, y, label='Seno')
plt.title('Gráfica con anotación')
plt.xlabel('Tiempo')
plt.ylabel('Valor')
plt.legend()
plt.annotate('Pico local', xy=(1.57, 1), xytext=(2.5, 1.2),
             arrowprops=dict(arrowstyle='->'))
plt.grid(True)
plt.show()
````

````python
D1. SCATTER SIMPLE
plt.figure(figsize=(8,5))
plt.scatter(x_scatter, y_scatter)
plt.title('Scatter Plot simple')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
D2. SCATTER CON TAMAÑO Y COLOR
sizes = np.random.randint(20, 200, 100)
colors = np.random.rand(100)

plt.figure(figsize=(8,5))
plt.scatter(x_scatter, y_scatter, s=sizes, c=colors)
plt.title('Scatter con tamaño y color')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
D3. SCATTER CON MAPA DE COLOR
plt.figure(figsize=(8,5))
sc = plt.scatter(x_scatter, y_scatter, c=colors, cmap='viridis')
plt.colorbar(sc)
plt.title('Scatter con mapa de color')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
D4. SCATTER CON LÍNEA DE REGRESIÓN
coef = np.polyfit(x_scatter, y_scatter, 1)
poly1d_fn = np.poly1d(coef)

plt.figure(figsize=(8,5))
plt.scatter(x_scatter, y_scatter)
plt.plot(x_scatter, poly1d_fn(x_scatter), color='red')
plt.title('Scatter con línea de regresión')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
D5. LÍNEA COMPARATIVA DE DOS VARIABLES
plt.figure(figsize=(8,5))
plt.plot(x, y, label='Seno')
plt.plot(x, y2, label='Coseno')
plt.title('Línea comparativa')
plt.legend()
plt.grid(True)
plt.show()
D6. REGRESIÓN CUADRÁTICA
coef2 = np.polyfit(x_scatter, y_scatter, 2)
poly2d_fn = np.poly1d(coef2)

x_sorted = np.sort(x_scatter)

plt.figure(figsize=(8,5))
plt.scatter(x_scatter, y_scatter)
plt.plot(x_sorted, poly2d_fn(x_sorted), color='green')
plt.title('Regresión cuadrática')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
D7. HEATMAP BÁSICO CON imshow()
matriz = np.random.rand(5, 5)

plt.figure(figsize=(6,5))
plt.imshow(matriz)
plt.title('Heatmap básico')
plt.colorbar()
plt.show()
D8. HEATMAP PERSONALIZADO
plt.figure(figsize=(6,5))
plt.imshow(matriz, cmap='coolwarm', interpolation='nearest')
plt.title('Heatmap personalizado')
plt.xticks(range(5))
plt.yticks(range(5))
plt.colorbar()
plt.show()
D9. HEATMAP DE MATRIZ DE CORRELACIÓN
df_corr = pd.DataFrame({
    'Satisfaccion': np.random.randint(50, 100, 50),
    'Productividad': np.random.randint(40, 100, 50),
    'Burnout': np.random.randint(20, 90, 50),
    'Horas': np.random.randint(30, 60, 50)
})

corr = df_corr.corr()

plt.figure(figsize=(6,5))
plt.imshow(corr, cmap='viridis')
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title('Heatmap de correlación')
plt.colorbar()
plt.show()
D10. HEATMAP CON ANOTACIONES NUMÉRICAS
plt.figure(figsize=(6,5))
plt.imshow(corr, cmap='coolwarm')
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title('Heatmap con anotaciones')

for i in range(len(corr)):
    for j in range(len(corr.columns)):
        plt.text(j, i, f"{corr.iloc[i, j]:.2f}", ha='center', va='center', color='black')

plt.colorbar()
plt.show()
````
