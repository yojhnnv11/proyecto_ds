## Objetivo
Explorar la creación y manipulación de Series en Pandas para comprender su utilidad en el análisis de datos.

```python
import pandas as pd
import numpy as np
```

```python
serie_numeros = pd.Series(range(1,11))

serie_numeros
```

```python
# Crear una Serie con materias favoritas

materias = pd.Series([
    "Programación",
    "Matemáticas",
    "Base de Datos",
    "Inteligencia Artificial",
    "Estadística"
])

materias
```

```python
# Crear una Serie con precios

precios = pd.Series([2500,3700,5900,1200,4800])

print(precios)

print("Tipo de dato:", precios.dtype)
```

```python
# Serie con valores booleanos

booleanos = pd.Series([True, False, True, True, False])

booleanos
```

```python
# Mostrar los tres primeros valores

serie = pd.Series([5,10,15,20,25,30])

serie.head(3)
```

```python
# Crear una Serie desde un diccionario

serie_dic = pd.Series({
    'a':10,
    'b':20,
    'c':30
})

serie_dic
```

```python
# Serie con notas de materias

notas = pd.Series({
    'Matemáticas':4.5,
    'Inglés':3.8,
    'Programación':5.0
})

notas
```

```python
# Serie de temperaturas

temperaturas = pd.Series({
    'Lunes':30,
    'Martes':28,
    'Miércoles':27
})

print(temperaturas)

print("Índices:", temperaturas.index)
```

```python
# Crear un array aleatorio

array = np.random.randint(0,100,10)

serie_array = pd.Series(array)

serie_array
```

```python
# Serie del 1 al 10 usando NumPy

serie = pd.Series(np.arange(1,11))

serie
```

```python
# Conversión de temperaturas

celsius = pd.Series([0,10,20,30,40])

fahrenheit = (celsius * 9/5) + 32

fahrenheit
```

```python
# Crear serie constante

serie_constante = pd.Series(100, index=range(5))

serie_constante
```

```python
# Serie con índices personalizados

personas = pd.Series(
    ['Juan','Ana','Luis'],
    index=['ID1','ID2','ID3']
)

personas
```

```python
personas['ID2']
```

```python
serie = pd.Series([10,20,30,40,50])

print("Index:", serie.index)
print("Valores:", serie.values)
print("Tipo:", serie.dtype)
```

```python
serie = pd.Series(range(10))

serie.iloc[2]
```

```python
serie = pd.Series(range(10), index=list("abcdefghij"))

serie.loc['d']
```

```python
s1 = pd.Series([1,2,3])
s2 = pd.Series([4,5,6])

s1 + s2
```

```python
s1 * 2
```

```python
serie = pd.Series([1,2,3,4])

serie.apply(lambda x: x**2)
```

```python
serie = pd.Series([10,25,60,80,45])

serie[serie > 50]
```

```python
serie = pd.Series([1,2,np.nan,4])

serie.fillna(0)
```

## Reflexión

El uso de Series en Pandas permite organizar y analizar datos de manera eficiente dentro de proyectos de análisis de datos.

Las Series funcionan como estructuras unidimensionales que facilitan operaciones matemáticas, filtrado, ordenamiento y transformación de datos. Gracias a sus índices personalizados y su integración con NumPy, es posible trabajar con grandes volúmenes de información de manera clara y estructurada.

Esto mejora la comprensión de los datos y optimiza el proceso de análisis dentro de proyectos de ciencia de datos.
